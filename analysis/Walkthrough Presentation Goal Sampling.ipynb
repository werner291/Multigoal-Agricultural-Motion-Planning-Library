# Introduction

To a robot, fruit trees represent a highly-complex obstacle, featuring numerous branches and thin twigs that must not be collided with.

In [[COMPAG paper|previous work]], we have explored the idea of a quadcopter, equipped with a robotic arm, bringing the end-effector of the arm up close to many fruits within such fruit trees, and demonstrated that the use of a [[Convex Hull]] around the canopy of the tree provided an effective means of traveling around the global outside of the tree.

However, in that work, a significant bottleneck was always reaching the deeper, innermost fruits within the tree canopy. A significant part of the runtime was spent attempting to reach these fruit, or even simply on ruling out their reachability.

Indeed, if we just look at the tree, we see that the number of tiny branches twigs is significant in our tree models:

```python
tree_meshes = pymgodpl.load_tree_meshes("appletree")
```

```python
viewer = pymgodpl.SimpleVtkViewer(True)
viewer.add_tree(tree_meshes, False, True)
viewer.position_camera(Vec3d(4,4,4),Vec3d(0,0,2.1))
_ = plt.imshow(viewer.show())
```
